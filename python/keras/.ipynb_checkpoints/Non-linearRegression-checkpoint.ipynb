```python
import keras
from keras.src.layers import *
from keras.src.optimizers import *
from keras.src.losses import *
from keras.src.activations import *
from keras.src.metrics import *
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
x = np.linspace(-10, 10, 75)
n = np.random.normal(0, 5, x.shape)
y = x ** 2 + n
# y = (np.e ** x - np.e ** -x) / (np.e ** x + np.e ** -x) 

plt.scatter(x, y, c='r')
```

```python
model = keras.Sequential([
    Input(shape=(1, )), 
    Dense(units=64, activation=tanh), 
    Dense(units=64, activation=tanh), 
    Dense(units=1), 
])
model.compile(
    optimizer=Adam(), 
    loss=mean_squared_error, 
    metrics=Accuracy
)
```

```python
model.summary()
model.fit(x, y, epochs=200)
```

```python
plt.scatter(x, y, c='r')
plt.plot(x, model.predict(x))
```
